```python
from Make_logistic import *
import seaborn as sns
from sklearn.model_selection import StratifiedKFold
import numpy as np
```

```python
df = pd.read_excel('../Data/Data_all_time.xlsx', index_col=0)
X,y=df.loc[:,['acetald_season_m18', 'acetone_season_m18', 'form_season_m18',
       'pm25_season_m18', 'nox_season_m18', 'bc_season_m18', 'no2_season_m18',
       'lrti_1y', 'genetic_risk_score_asthma', 'gestational_age',
       'no2_str_m18', 'nox_str_m18', 'pm25_str_m18']],df.verificeret_18y
```

```python
y
```

```python
roc=[]
CV_repeats=2
f1=[]
roc_train=[]
f1_train=[]
for i, CV_repeat in enumerate(range(CV_repeats)): 
    #Verbose 
    print('\n------------ CV Repeat number:', CV_repeat)
    #Establish CV scheme
    CV =StratifiedKFold(n_splits=3, random_state=None, shuffle=True) # Set random state 
    ix_training, ix_test = [], []
    # Loop through each fold and append the training & test indices to the empty lists above
    for fold in CV.split(X,y):
        ix_training.append(fold[0]), ix_test.append(fold[1])
        
    ## Loop through each outer fold and extract SHAP values 
    for i, (train_outer_ix, test_outer_ix) in enumerate(zip(ix_training, ix_test)): 
        #Verbose
        print('\n------ Fold Number:',i)
        X_train, X_test = torch.tensor(X.iloc[train_outer_ix, :].to_numpy()).float(), torch.tensor(X.iloc[test_outer_ix, :].to_numpy()).float()
        y_train, y_test = torch.tensor(y.iloc[train_outer_ix].to_numpy()).float(), torch.tensor(y.iloc[test_outer_ix].to_numpy()).float()
        mcmc,kernel=run_inference(features=X_train, target=y_train, N_warmup_steps=3000 , number_of_chains=1)
        posterior_samples = mcmc.get_samples(500)
        predictive= Predictive(logistic_regression_model, posterior_samples)
        posterior_predictive=predictive((X_train))
        prior = Predictive(logistic_regression_model, num_samples=500)(X_train)
        pyro_data = az.from_pyro(mcmc,prior=prior,posterior_predictive=posterior_predictive)
        az.to_json(pyro_data, "asthma_18_year_"+str(i)+str(CV_repeat)+".json")
        roc_train.append(roc_auc_score(y_train, predictive(X_train)["obs"].mean(axis=0)))
        f1_train.append(f1_score(y_train,kernel(X_train)))
        roc.append(roc_auc_score(y_test, predictive(X_test)["obs"].mean(axis=0)))
        f1.append(f1_score(y_test,kernel(X_test)))
```

```python
np.mean(roc),np.std(roc),np.mean(f1),np.std(f1)
```

```python
np.mean(roc_train),np.std(roc_train),np.mean(f1_train),np.std(f1_train)
```

```python
asthma_1_exposure=az.from_json("asthma_18_year_00.json")
asthma_2_exposure=az.from_json("asthma_18_year_01.json")
asthma_3_exposure=az.from_json("asthma_18_year_10.json")
asthma_4_exposure=az.from_json("asthma_18_year_11.json")
asthma_5_exposure=az.from_json("asthma_18_year_20.json")
asthma_6_exposure=az.from_json("asthma_18_year_21.json")
asthma=az.concat(asthma_1_exposure,asthma_2_exposure,asthma_3_exposure,asthma_4_exposure,asthma_5_exposure,asthma_6_exposure, dim="chain")
```

```python
asthma["posterior"]["weight"]=np.exp(asthma["posterior"]["weight"])
list_parameters=["Indoor Acetald","Indoor Acetone","Indoor formaldehyd","Indoor $PM_{2.5}$","Indoor $NO_x$","Indoor black Carbon","Indoor $NO_2$","Lower respiratory tract infection","Genetics","Gestational age","Outdoor $NO_2$","Outdoor $NO_x$","Outdoor $PM_{2.5}$"]

asthma.posterior = asthma.posterior.assign_coords(
    weight_dim_1=list_parameters
)
asthma.to_netcdf("asthma_18y.nc")
```
