```python
param_distributions = {
    'n_estimators': randint(100, 500),
    'max_depth': randint(4, 10),"min_impurity_decrease":uniform(0.001, 0.03),"min_samples_leaf":[5]
,"class_weight":["balanced"]}
```

```python
def nested_cross_validation(X,y,model,N_CV,N_split,para):
    shap_values_per_cv=np.zeros((X.shape[0], X.shape[1], 2))
    performence_test=[]
    performence_train=[]
    f1_test=[]
    f1_train=[]
    feature=[]
    parameters=[]
    j=0
    for i, CV_repeat in enumerate(range(N_CV)): 
        #Verbose 
        #Establish CV scheme
        if model==RandomForestClassifier():
            CV =StratifiedKFold(n_splits=N_split, random_state=i, shuffle=True) # Set random state 
        ## Loop through each outer fold and extract SHAP values 
            for i, (train_outer_ix, test_outer_ix) in enumerate(CV.split(X,y)): 
                #Verbose
                j=j+1
                
                X_train, X_test = X.iloc[train_outer_ix, :], X.iloc[test_outer_ix, :]
                y_train, y_test = y.iloc[train_outer_ix], y.iloc[test_outer_ix]
    
                ## Establish inner CV for parameter optimization #-#-#
                cv_inner = StratifiedKFold(n_splits=N_split, random_state=i, shuffle=True) #-#-#
                search = RandomizedSearchCV(estimator=model,param_distributions=para, cv=cv_inner,scoring="roc_auc",random_state=42,  n_jobs=-1,verbose=1)
                result = search.fit(X_train, y_train) #-#=#
                # Fit model on training data 
                result.best_estimator_.fit(X_train, y_train) #-#-#
                # Use SHAP to explain predictions using best estimator 
                explainer = shap.TreeExplainer(result.best_estimator_) 
                shap_values = explainer.shap_values(X_test)
                shap_values_per_cv[test_outer_ix,:,:]+=shap_values
                performence_test.append(result.best_estimator_.score(X_test,y_test))
                performence_train.append(result.best_estimator_.score(X_train,y_train))
                f1_test.append(f1_score(y_test,result.best_estimator_.predict(X_test)))
                f1_train.append(f1_score(y_train,result.best_estimator_.predict(X_train)))
                feature.append(result.best_estimator_.feature_importances_)
                parameters.append(result.best_params_)
                print('done %',(j/6))
                total_shap=shap_values_per_cv/N_CV
                return peformence=pd.DataFrame([performence_train,performence_test,f1_train,f1_test],index=["AUROC_train","AUROC_test","F1_train","F1_test"])
            else:
                CV =KFold(n_splits=N_split, random_state=i, shuffle=True) # Set random state 
            ## Loop through each outer fold and extract SHAP values 
                for i, (train_outer_ix, test_outer_ix) in enumerate(CV.split(X)): 
                    #Verbose
                    j=j+1
                    
                    X_train, X_test = X.iloc[train_outer_ix, :], X.iloc[test_outer_ix, :]
                    y_train, y_test = y.iloc[train_outer_ix], y.iloc[test_outer_ix]
        
                    ## Establish inner CV for parameter optimization #-#-#
                    cv_inner = KFold(n_splits=N_split, random_state=i, shuffle=True) #-#-#
                    search = RandomizedSearchCV(estimator=model,param_distributions=para, cv=cv_inner,random_state=42,  n_jobs=-1,verbose=1)
                    result = search.fit(X_train, y_train) #-#=#
                    # Fit model on training data 
                    result.best_estimator_.fit(X_train, y_train) #-#-#
                    # Use SHAP to explain predictions using best estimator 
                    explainer = shap.TreeExplainer(result.best_estimator_) 
                    shap_values = explainer.shap_values(X_test)
                    shap_values_per_cv[test_outer_ix,:,:]+=shap_values
                    performence_test.append(result.best_estimator_.score(X_test,y_test))
                    performence_train.append(result.best_estimator_.score(X_train,y_train))
                    f1_test.append(f1_score(y_test,result.best_estimator_.predict(X_test)))
                    f1_train.append(f1_score(y_train,result.best_estimator_.predict(X_train)))
                    feature.append(result.best_estimator_.feature_importances_)
                    parameters.append(result.best_params_)
                    print('done %',(j/6))
                    total_shap=shap_values_per_cv/N_CV
                    return peformence=pd.DataFrame([performence_train,performence_test,f1_train,f1_test],index=["AUROC_train","AUROC_test","F1_train","F1_test"])

    return peformence.T,total_shap,pd.DataFrame(feature,columns=X.columns).mean(),parameters
```
