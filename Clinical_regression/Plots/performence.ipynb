```python
x=np.array([1,2,3])
```

```python
fig,ax=plt.subplots(2,2,figsize=(10,10))
ax[0,0].set_title('AUROC unadjusted')

ax[0,0].errorbar(x-0.1,roc_test_auroc_mean,yerr=roc_test_auroc_std,fmt='Db',label="Test Logistic")
ax[0,0].errorbar(x-0.05,roc_train_auroc_mean,yerr=roc_train_auroc_std,fmt='.b',label="Train Logistic")
ax[0,0].errorbar(x+0.05,test_auroc_mean_rf,yerr=test_auroc_std_rf,fmt='dr',label="Test Random forest")
ax[0,0].errorbar(x+0.1,train_auroc_mean_rf,yerr=train_auroc_std_rf,fmt='.r',label="Train Random forest")
ax[0,0].plot(np.arange(0,5),np.ones(5)*0.6,"k")
ax[0,0].plot(np.arange(0,5),np.ones(5)*0.7,"k")
ax[0,0].legend(loc=1)
ax[0,0].set_ylim(0.45,1)
ax[0,0].set_xticks(x,["7 Years","12 Years","18 Years"])
ax[0,1].set_title('F1 unadjusted')
ax[0,1].errorbar(x-0.1,test_F1_mean,yerr=test_F1_std,fmt='db',label="Test Logistic")
ax[0,1].errorbar(x-0.05,train_F1_mean,yerr=train_F1_std,fmt='.b',label="Train Logistic")
ax[0,1].errorbar(x+0.05,test_F1_mean_rf,yerr=test_F1_std_rf,fmt='dr',label="Test Random forest")
ax[0,1].errorbar(x+0.1,train_F1_mean_rf,yerr=train_F1_std_rf,fmt='.r',label="Train Random forest")
ax[0,1].plot(np.arange(0,5),np.ones(5)*0.2,"k")
ax[0,1].plot(np.arange(0,5),np.ones(5)*0.3,"k")
ax[0,1].legend(loc=1)
ax[0,1].set_ylim(0,0.9)
ax[0,1].set_xticks(x,["7 Years","12 Years","18 Years"])
ax[1,0].set_title('AUROC adjusted')
ax[1,0].errorbar(x-0.1,roc_test_auroc_mean_adjust,yerr=roc_test_auroc_std_adjust,fmt='db',label="Test Logistic")
ax[1,0].errorbar(x-0.05,roc_train_auroc_mean_adjust,yerr=roc_train_auroc_std_adjust,fmt='.b',label="Train Logistic")
ax[1,0].errorbar(x+0.05,test_auroc_mean_rf,yerr=test_auroc_std_rf,fmt='dr',label="Test Random forest")
ax[1,0].errorbar(x+0.1,train_auroc_mean_rf,yerr=train_auroc_std_rf,fmt='.r',label="Train Random forest")
ax[1,0].plot(np.arange(0,5),np.ones(5)*0.6,"k")
ax[1,0].plot(np.arange(0,5),np.ones(5)*0.7,"k")
ax[1,0].legend(loc=1)
ax[1,0].set_ylim(0.45,1)
ax[1,0].set_xticks(x,["7 Years","12 Years","18 Years"])
ax[1,1].set_title('F1 adjusted')
ax[1,1].errorbar(x-0.1,test_F1_mean_rf_adjust,yerr=test_F1_std_rf_adjust,fmt='db',label="Test Logistic")
ax[1,1].errorbar(x-0.05,train_F1_mean_adjust,yerr=train_F1_std_adjust,fmt='.b',label="Train Logistic")
ax[1,1].errorbar(x+0.05,test_F1_mean_rf_adjust,yerr=test_F1_std_rf_adjust,fmt='dr',label="Test Random forest")
ax[1,1].errorbar(x+0.1,train_F1_mean_rf_adjust,yerr=train_F1_std_rf_adjust,fmt='r.',label="Train Random forest")
ax[1,1].plot(np.arange(0,5),np.ones(5)*0.2,"k")
ax[1,1].plot(np.arange(0,5),np.ones(5)*0.3,"k")
ax[1,1].legend(loc=1)
ax[1,1].set_ylim(0,0.9)

ax[1,1].set_xticks(x,["7 Years","12 Years","18 Years"])
fig.savefig("performence.png" ,)
```
