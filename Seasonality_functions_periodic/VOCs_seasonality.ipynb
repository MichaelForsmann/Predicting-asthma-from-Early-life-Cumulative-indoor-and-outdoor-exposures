```python
data=pd.read_excel(r"../Data_and_Preprocessing/before_season_correction.xlsx")
```

```python
train=data.loc[data.label==0,:]
```

```python
train_form=train.dropna(subset="formaldehyd")
X_formaldehyd,y_formaldehyd=torch.tensor(train_form.corrected_week.values).to(device),torch.tensor(train_form.formaldehyd.values).to(device)
nuts_kernel_formaldehyd,gpr_formaldehyd=model(X_formaldehyd,y_formaldehyd,device) 
pyro_form,gpr_formaldehyd=train_model(X_formaldehyd,nuts_kernel_formaldehyd,gpr_formaldehyd,device,steps_num=2000) 
torch.save(gpr_formaldehyd,"models/formaldehyd_seasonality");
az.to_json(pyro_form, "Arviz_stats/mcmc_form_seasonality.json")
```

```python
train_acetald=train.dropna(subset="acetald")
X_acetald,y_acetald=torch.tensor(train_acetald.corrected_week.values).to(device),torch.tensor(train_acetald.acetald.values).to(device)
nuts_kernel_formaldehyd,gpr_formaldehyd=model(X_acetald,y_acetald,device) 
nuts_kernel_acetald,gpr_acetald=model(X_acetald,y_acetald,device) 
pyro_acetald,gpr_acetald=train_model(X_acetald,nuts_kernel_acetald,gpr_acetald,device,steps_num=2000) 
torch.save(gpr_acetald,"models/acetald_seasonality");
az.to_json(pyro_acetald, "Arviz_stats/acetald_seasonality.json")
```

```python
train_acetone=train.dropna(subset="acetone")
X_acetone,y_acetone=torch.tensor(train_acetone.corrected_week.values).to(device),torch.tensor(train_acetone.acetone.values).to(device)
nuts_kernel_acetone,gpr_acetone=model(X_acetone,y_acetone,device) 
pyro_acetone,gpr_acetone=train_model(X_acetone,nuts_kernel_acetone,gpr_acetone,device,steps_num=2000) 
torch.save(gpr_acetone,"models/acetone_seasonality");
az.to_json(pyro_acetone, "Arviz_stats/mcmc_acetone_seasonality.json")
```

```python
train_voc=train.dropna(subset="Total_VOCs")
X_vocs,y_vocs=torch.tensor(train_voc.corrected_week.values).to(device),torch.tensor(train_voc.Total_VOCs.values).to(device)

nuts_kernel_vocs,gpr_vocs=model(X_vocs,y_vocs,device) 
pyro_vocs,gpr_vocs=train_model(X_vocs,nuts_kernel_vocs,gpr_vocs,device,steps_num=2000) 
torch.save(gpr_vocs,"models/vocs_seasonality");
az.to_json(pyro_vocs, "Arviz_stats/mcmc_vocs_seasonality.json")
```
