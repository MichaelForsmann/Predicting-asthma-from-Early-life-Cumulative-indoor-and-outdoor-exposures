```python
data=pd.read_excel(r"../Data_and_Preprocessing/before_season_correction.xlsx")
device="cuda"
train=data.loc[data.label==0,:]
```

```python
import torch.multiprocessing as mp
if __name__ == '__main__':
    mp.set_start_method('spawn', force=True)
```

```python
train_no2=train.dropna(subset="no2")
X_no2,y_no2=torch.tensor(train_no2.corrected_week.values).to(device),torch.tensor(train_no2.no2.values).to(device)
nuts_kernel,gpr_no2=model(X_no2,y_no2,device) 
pyro_data_no2,gpr_no2=train_model(X_no2,nuts_kernel,gpr_no2,device,steps_num=2000) 
az.to_json(pyro_data_no2, "Arviz_stats/mcmc_no2_seasonality.json")
torch.save(gpr_no2,"models/no2_seasonality");
```

```python
train_nox=train.dropna(subset="nox")
X_nox,y_nox=torch.tensor(train_nox.corrected_week.values).to(device),torch.tensor(train_nox.nox.values).to(device)
nuts_kernel,gpr_nox=model(X_nox,y_nox,device) 
pyro_data_nox,gpr_nox=train_model(X_nox,nuts_kernel,gpr_nox,device,steps_num=2000) 
az.to_json(pyro_data_nox, "Arviz_stats/mcmc_nox_seasonality.json")
torch.save(gpr_nox,"models/nox_seasonality");
```
