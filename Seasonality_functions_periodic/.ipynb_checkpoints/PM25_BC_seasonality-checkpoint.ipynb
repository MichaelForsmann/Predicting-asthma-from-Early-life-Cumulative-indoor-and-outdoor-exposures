```python
def model(X, y, device, sigma_prior=None):
    clear_param_store()
    
    if sigma_prior is None:
        sigma_prior = y.std().to(device)

    # Periodic Kernel: Sees Dim 0
    # input_dim should be total dims (3)
    periodic = gp.kernels.Periodic(input_dim=1, active_dims=[0])
    periodic.period = torch.tensor(float(X[:, 0].max()), device=device)
    periodic.variance = PyroSample(dist.HalfNormal(torch.tensor(y.mean().cpu().detach().numpy(), device=device)))
    periodic.lengthscale = PyroSample(dist.HalfNormal(torch.tensor(2., device=device)))

    # RBF Kernel: Sees Dim 1 and 2
    rbf = gp.kernels.RBF(input_dim=1, active_dims=[1])
    
    # FIX 1: Variance must be a scalar
    rbf.variance = PyroSample(dist.HalfNormal(torch.tensor(y.mean().cpu().detach().numpy(), device=device)))
    
    # FIX 2: Use .to_event(1) for the vector lengthscale
    rbf.lengthscale = PyroSample(dist.HalfNormal(torch.tensor(1., device=device)))
    rbf1 = gp.kernels.Linear(input_dim=1, active_dims=[2])
    
    # FIX 1: Variance must be a scalar
    rbf1.variance = PyroSample(dist.HalfNormal(torch.tensor(y.mean().cpu().detach().numpy(), device=device)))
    rbf2 = gp.kernels.Linear(input_dim=1, active_dims=[3])

    # FIX 1: Variance must be a scala
    rbf2.variance = PyroSample(dist.HalfNormal(torch.tensor(y.mean().cpu().detach().numpy(), device=device)))
    rbf3 = gp.kernels.Linear(input_dim=1, active_dims=[4])

    rbf3.variance = PyroSample(dist.HalfNormal(torch.tensor(y.mean().cpu().detach().numpy(), device=device)))

    # FIX 2: Use .to_event(1) for the vector lengthscale
    # Combine
    addition =  gp.kernels.Sum(gp.kernels.Sum(periodic, rbf),rbf1).to(device)
    interaction=gp.kernels.Product(periodic,rbf2)
    interaction1=gp.kernels.Product(interaction,rbf3)
    combined=gp.kernels.Sum(interaction1,addition)
    
    gpr = gp.models.GPRegression(X, y, combined).to(device)
    gpr.noise = PyroSample(dist.HalfNormal(sigma_prior))
    
    return NUTS(gpr.model), gpr
```

```python
train_pm25=train.dropna(subset="pm_final")
test_pm25=test.dropna(subset="pm_final")
X_pm25_test,y_pm25_test=torch.from_numpy(test_pm25.loc[:,["corrected_week","flow","Area","smoker"]].values).to(device),torch.tensor(test_pm25.pm_final.values).to(device)
X_pm25,y_pm25=torch.from_numpy(train_pm25.loc[:,["corrected_week","flow","Area","smoker"]].values).to(device),torch.tensor(train_pm25.pm_final.values).to(device)
nuts_kernel,gpr_pm25=model(X_pm25,y_pm25,device) 
pyro_data_pm25,gpr_pm25=train_model(X_pm25,nuts_kernel,gpr_pm25,device,steps_num=5000) 
```
